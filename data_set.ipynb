```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ... (your previous code) ...

# Define discrete_features - Replace with your actual discrete feature names
discrete_features = ['cnt', 't1', 't2', 'hum', 'wind_speed', 'weather_code', 'is_holiday', 'is_weekend', 'season']

def plot_feature_distribution(features, data):
    # Calculate the number of rows and columns for subplots
    num_features = len(features)
    num_cols = 2  # You can adjust this
    num_rows = (num_features + num_cols - 1) // num_cols  # Calculate rows needed

    fig, axes = plt.subplots(num_rows, num_cols, figsize=(12, 10))  # Create subplots with calculated dimensions
    axes = axes.flatten()  # Flatten for easier iteration

    for i, feature in enumerate(features):
        if feature in data.columns:
            sns.histplot(data[feature], ax=axes[i])
            axes[i].set_title(f'Distribution of {feature}')
            axes[i].set_xlabel(feature)
            axes[i].set_ylabel('Count')
        else:
            print(f"Column '{feature}' not found in DataFrame. Skipping...")

    # Hide any unused subplots (if num_features is not a multiple of num_cols * num_rows)
    for i in range(num_features, len(axes)):
        axes[i].set_visible(False)

    plt.tight_layout()
    plt.show()

plot_feature_distribution(discrete_features, df)
```

```python

df['year'] = df['timestamp'].dt.year
df['month'] = df['timestamp'].dt.month
df['day_of_month'] = df['timestamp'].dt.day
df['day_of_week'] = df['timestamp'].dt.dayofweek  # Monday=0, Sunday=6
df['hour'] = df['timestamp'].dt.hour

def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Autumn'

df['season'] = df['month'].apply(get_season)

print(df.head())
```

```python
import seaborn as sns

import matplotlib.pyplot as plt

import pandas as pd

# Example dataset (replace with your actual data)

data = {

    'timestamp': pd.date_range(start='2020-01-01', periods=600, freq='D'),  # 600 days starting from Jan 1, 2020

    'bike_shares': [100 + (x*5) for x in range(600)],  # Example bike shares data (increasing over time)

    'holiday': [1 if x % 15 == 0 else 0 for x in range(600)],  # Simulate holidays every 15th day

    'season': ['Winter', 'Spring', 'Summer', 'Fall'] * 150  # Simulated seasons for simplicity

}

# Create DataFrame

df = pd.DataFrame(data)

# Extract year from the timestamp

df['year'] = df['timestamp'].dt.year

# 1. Plotting Bike Shares by Year

plt.figure(figsize=(10, 6))

# Group by year and calculate the total bike shares for each year

df_yearly = df.groupby('year')['bike_shares'].sum().reset_index()

# Plotting the data

sns.barplot(x='year', y='bike_shares', data=df_yearly, palette='viridis')

plt.title('Total Bike Shares by Year')

plt.xlabel('Year')

plt.ylabel('Total Bike Shares')

# Show plot

plt.tight_layout()

plt.show()

# 2. Plotting Bike Shares on Holidays by Season

plt.figure(figsize=(10, 6))

# Filter data for holidays

df_holidays = df[df['holiday'] == 1]

# Group by season and calculate total bike shares on holidays

df_season_holiday = df_holidays.groupby('season')['bike_shares'].sum().reset_index()

# Plotting the data

sns.barplot(x='season', y='bike_shares', data=df_season_holiday, palette='Set2')

plt.title('Bike Shares on Holidays by Season')

plt.xlabel('Season')

plt.ylabel('Total Bike Shares on Holidays')

# Show plot

plt.tight_layout()

plt.show()

```
